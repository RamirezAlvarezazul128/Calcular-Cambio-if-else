Problema 1: La suma de un número más 9 es igual a 21 Planteamiento del problema:Llamaremos $x$ al número desconocido.La frase "la suma de un número más 9" se traduce como: $x + 9$."Es igual a 21" completa nuestra ecuación.Ecuación:$$x + 9 = 21$$Despeje de la variable:Para dejar a la $x$ sola, el 9 que está sumando pasa al otro lado de la igualdad restando:$$x = 21 - 9$$Realizamos la resta:$$x = 12$$Solución: El número buscado es 12.

```python
# Problema 1: Encontrar el número desconocido
# Ecuación: x + 9 = 21 -> x = 21 - 9
# numero_conocido = 9
resultado_igualdad = 21
numero_conocido = 9
resultado_total=int(input("ingrese el resultado total: "))
numero_conocido=int(input("ingrese el numero conocido : "))
# Despejamos x
x = resultado_igualdad - numero_conocido

print(f"El número buscado en el problema 1 es: {x}")
```

Problema 2: Edades de Pedro y Ana
Planteamiento del problema:Edad de Ana: La definiremos como $x$ (ya que es la referencia más sencilla).Edad de Pedro: El problema dice que tiene el "doble de edad que su hermana", por lo tanto es $2x$.La suma de ambas edades es 36.Ecuación:$$x + 2x = 36$$Despeje de la variable:Primero, sumamos las términos semejantes ($1x + 2x$):$$3x = 36$$Para despejar $x$, el 3 que está multiplicando pasa al otro lado dividiendo:$$x = \frac{36}{3}$$Realizamos la división:$$x = 12$$Solución:La edad de Ana ($x$) es 12 años.La edad de Pedro ($2x$) es $2 \cdot 12 =$ 24 años.
Verificación: $12 + 24 = 36$.

```python
# Problema 2: Edades de Pedro (2x) y Ana (x)
# Ecuación: x + 2x = 36 -> 3x = 36 -> x = 36 / 3
# suma_total = 36
suma_total = int(input("ingrese la suma total: "))
coeficiente_total = 1 + 2 # Ana (1) + Pedro (2)

# Calculamos la edad de Ana (x)
edad_ana = suma_total / coeficiente_total
# Calculamos la edad de Pedro (2x)
edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
